#unifying the dataset

```python
import os
import shutil
import uuid

class_mapping = {
    'Healthy': ['Healthy'],
    'Mosaic': ['Mosaic'],
    'RedRot': ['RedRot'],
    'Rust': ['Rust', 'BrownRust', 'RedRust'],
    'Yellow_Leaf': ['Yellow', 'Yellow Leaf'],
    'Banded_Chlorosis': ['Banded Chlorosis', 'Banded Chl...'],
    'Brown_Spot': ['Brown Spot'],
    'Grassy_Shoot': ['Grassy shoot'],
    'Pokkah_Boeng': ['Pokkah Boeng'],
    'Sett_Rot': ['Sett Rot'],
    'Viral_Disease': ['Viral Disease'],
    'Smut': ['smut']
}

source_datasets = ['dataset_1', 'dataset_2', 'dataset_3', 'dataset_4']
target_dir = 'datasets'
os.makedirs(target_dir, exist_ok=True)

total_moved = 0
for target_class, source_folders in class_mapping.items():
    target_class_path = os.path.join(target_dir, target_class)
    os.makedirs(target_class_path, exist_ok=True)
    for dataset in source_datasets:
        for source_folder in source_folders:
            source_path = os.path.join(dataset, source_folder)
            if os.path.exists(source_path):
                files = os.listdir(source_path)
                for filename in files:
                    file_path = os.path.join(source_path, filename)
                    if os.path.isfile(file_path):
                        ext = os.path.splitext(filename)[1]
                        base_name = os.path.splitext(filename)[0]
                        new_filename = f"{base_name}_{dataset}_{uuid.uuid4().hex[:6]}{ext}"
                        dest_path = os.path.join(target_class_path, new_filename)
                        shutil.move(file_path, dest_path)
                        total_moved += 1

print(f"{total_moved} The images were successfully moved to the folder. '{target_dir}'.")
```
